Hier wird $\sin^2(\delta_l)$ gegen $y=k\cdot r_0$ aufgetragen

```python
def plot_sin2_vs_energy(l, v0_list):
    '''Plot des Sinusquadrats der Streuphase für auswählbare Potentialtiefen einer bestimmten Ordnung'''
    plt.figure(figsize=(10,6))
    for v0 in v0_list:
        x = np.linspace(0.0001, 50, 10000000)     # Start bei 0.0001 wegen log-Achse
        a = rs.numerator(x, v0, l) / rs.denominator(x, v0, l)
        y = 1 / (1 + 1/a**2)               # Ausdruck für sin^2(delta)
        plt.plot(x, y, label=f"$v_0={v0:g}$")  

    plt.xscale("log")
    plt.yscale("log")
    plt.xlim(0.1, 50)
    plt.ylim(0.0001, 1.1)
    plt.xlabel(r'$y_R=k r_0$')
    plt.ylabel(rf'$\sin^2(\delta_{l})$')
    plt.grid(True, which="both", ls="--", alpha=0.3)
    plt.legend(title=f"$l= {l}$")
    plt.tight_layout()
    plt.show()

#Aufrufe für in der Plots zum Sinus der Streuphase in der Bachelorarbeit:
plot_sin2_vs_energy(0, [1, 2, np.pi**2/4, 5]) 
plot_sin2_vs_energy(4, [22.94, 30, 48])
```
